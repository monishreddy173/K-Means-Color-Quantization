```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('palm_trees.jpg')
```

```python
img
```

```python
img.shape
```

```python
plt.figure(dpi=200)
plt.imshow(img)
```

```python
(h,w,c)=img.shape
```

```python
img2d=img.reshape(h*w,c)
```

```python
img2d
```

```python
img
```

```python
len(img.shape)
```

```python
len(img2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(img2d)
```

```python
labels
```

```python
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_image=np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_image
```

```python
plt.figure(dpi=200)
plt.imshow(quantized_image)
```
